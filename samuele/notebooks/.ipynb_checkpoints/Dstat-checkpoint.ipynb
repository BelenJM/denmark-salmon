```python
def plot_Dstat(fileList, H3, H2, H1, indexOrder=None):
    idxName = H1
    idx = [ (i.split('.')[4]=='txt')&((i.split('.')[3]=='Observed')|(i.split('.')[3]=='ErrorCorr')) 
        for i in fileLists ]
    BUR_D = fileLists[ [(H3 in i)&(H2 in i)&(np.any([name in i for name in idxName])) for i in fileLists] ]
    BUR_obs = pd.DataFrame(index=idxName, columns=['D','SD','p'])
    BUR_err = pd.DataFrame(index=idxName, columns=['D','SD','p'])
    for F in BUR_D:
        X = pd.read_csv(f'{mypath1}/{F}', sep='\t')
        X2 = X[ X['H3']==H3 ]
        typeF = F.split('.')[3]
        print(typeF)
        print(X2)
        if( (np.array(X2['H2'])==H2)&(np.array(X2['H1']) in idxName) ):
            if(typeF=='Observed'):
                BUR_obs.loc[np.array(X2['H1']),:] = np.asarray(X2[ ['JK-D','V(JK-D)','pvalue'] ])
            else:
                BUR_err.loc[np.array(X2['H1']),:] = np.asarray(X2[ ['JK-D','V(JK-D)','pvalue'] ])
        if( (np.array(X2['H2'])!=H2)&(np.array(X2['H2']) in idxName) ):
            if(typeF=='ErrorCorr'):
                BUR_obs.loc[np.array(X2['H2']),:] = np.asarray(X2[ ['JK-D','V(JK-D)','pvalue'] ])
                BUR_obs.loc[np.array(X2['H2']),'D'] = BUR_obs.loc[np.array(X2['H2']),'D'] 
            else:
                BUR_err.loc[np.array(X2['H2']),:] = np.asarray(X2[ ['JK-D','V(JK-D)','pvalue'] ])
                BUR_err.loc[np.array(X2['H2']),'D'] = BUR_err.loc[np.array(X2['H2']),'D'] 
    
    BUR_df = pd.DataFrame(index = BUR_obs.index, columns=['Observed','Error-corrected','Population'])
    BUR_df['Observed'] = BUR_obs['D']
    BUR_df['Error-corrected'] = BUR_err['D']
    BUR_df['Observed_SD'] = BUR_obs['SD']
    BUR_df['Error-corrected_SD'] = BUR_err['SD']
    BUR_df['Population'] = BUR_df.index
    if indexOrder is None:
        BUR_df = BUR_df.sort_values("Observed", ascending=True)
    else:
        BUR_df = BUR_df.loc[indexOrder,:]
    import seaborn as sns

    # Make the PairGrid
    g = sns.PairGrid(BUR_df,
                     x_vars=['Observed','Error-corrected'], 
                     y_vars=["Population"],
                     height=10, aspect=.25)

    # Draw a dot plot using the stripplot function
    g.map(sns.stripplot, size=10, orient="h",
          palette="viridis", linewidth=1, edgecolor="w")

    # Use the same x axis limits on all columns and add better labels
    g.set(xlim=(-.3, +.3), xlabel="D-statistic", ylabel="")

    # Use semantically meaningful titles for the columns
    titles = ["Observed D-stat", 'Error-corrected D-stat']

    counter=0
    for ax, title in zip(g.axes.flat, titles):
        lineCoord=0
        # Set a different title for each axes
        ax.set(title=title)
        ax.axvline(0)
        # Make the grid horizontal instead of vertical
        ax.xaxis.grid(False)
        ax.yaxis.grid(True)
        for i in BUR_df.index:
            if counter==0:
                xVals = [ BUR_df.loc[i,'Observed']-3*np.sqrt(BUR_df.loc[i,'Observed_SD']), 
                      BUR_df.loc[i,'Observed']+3*np.sqrt(BUR_df.loc[i,'Observed_SD'])   ]
                ax.hlines(y=lineCoord, xmin=xVals[0], xmax=xVals[1], color='blue')
                #ax.axhline( y=lineCoord, xmin=xVals[0], xmax=xVals[1], color='red' )
                #print(xVals)
            if counter==1:
                xVals = [ BUR_df.loc[i,'Error-corrected']-3*np.sqrt(BUR_df.loc[i,'Error-corrected_SD']), 
                          BUR_df.loc[i,'Error-corrected']+3*np.sqrt(BUR_df.loc[i,'Error-corrected_SD'])   ]
                ax.hlines(y=lineCoord, xmin=xVals[0], xmax=xVals[1], color='red')
                #ax.axhline( y=lineCoord, xmin=xVals[0], xmax=xVals[1], color='red' )
            lineCoord += 1
    counter += 1
    print(BUR_df.index)
    return np.array(BUR_df.index)
    sns.despine(left=True, bottom=True)
```
